```python
data1 = data[["Date","Close"]]
data1.columns = ["ds" , "y"]
```

```python
data1.tail()
```

```python
sample = pd.read_csv("/kaggle/input/dataquest-abdous-company/sample_submission.csv")
long = sample.shape[0]
```

```python
model = NeuralProphet()
model.fit(df=data1 , freq="D", epochs=200)
```

```python
future = model.make_future_dataframe(data1,periods=520)
forecast = model.predict(future)
```

```python
forecast = forecast[forecast['ds'].isin(sample["Date"])]
```

```python
forecast
```

```python
forecast = forecast.loc[:,["ds","yhat1"]]
```

```python
forecast = forecast.rename(columns={'ds': 'Date', 'yhat1': 'Close'})
```
